```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('MumbaiHousing.csv')
df.head()
```

```python
df['price'] = df['price'].str.replace(',', '')
```

```python
df.head()
```

### Data Cleaning

```python
# df['price'].str.find('Cr')
df['price'].unique()
```

```python
units = []
for cols in df['price']:
#     print(cols[-2:])
    units.append(cols[-2:])
```

```python
df['units'] = units
```

```python
df
```

```python
df['price_sqft'] = df['price_sqft'].str.replace(',','')
df.head()
```

```python
df['price'] = df['price'].apply(lambda x: x.split(' ')[0])
df.head()
```

```python
df.info()
```

```python
df['price'].unique()
```

```python
df['price'] = df['price'].astype(float)
df['price_sqft'] = df['price_sqft'].astype(float)
```

```python
df.info()
```

```python
df['price'] = df.apply(lambda x: x['price']*100 if x['units'] == 'Cr' else x['price'],axis=1)
```

```python
df
```

```python
df.drop('units',inplace=True,axis=1)
```

```python
df['size'].unique()
```

```python
df['bhk'] = df['size'].apply(lambda x: x.split(' ')[0])
```

```python
df = df[df['bhk'] != 'R']
```

```python
df['bhk'] = df['bhk'].astype(int)
df.head()
```

```python
# Adding apartment type -> Info extracted from 'size'
df['type'] = df.apply(lambda x:x['size'].split(' ')[1] if len(x['size'].split(' ')) >1 else 'BHK', axis=1)
```

```python
# Adding housing type -> Info extracted from 'size'
df['housing_type'] = df.apply(lambda x:''.join(x['size'].split(' ')[2:]) if len(x['size'].split(' ')) >2 else 'Apartment', axis=1)
```

```python
df.housing_type.unique()
```

```python
df= df.drop('size',axis=1)
```

```python
df.head()
```

## EDA & Feature Extraction

```python
df['area_type'].value_counts()
```

```python
# Function to analyse the average price of a category
def get_mean_price(df,attribute):
    attr_map = {}
    for attr in df[attribute].unique():
        attr_map[attr] = df[df[attribute] == attr]['price'].mean()
    return attr_map
```

```python
area_type_average = get_mean_price(df,'area_type')
```

```python
area_type_average
```

#### Not much different in prices wrt area_type 

```python
availability_average = get_mean_price(df,'availability')
availability_average
```

#### Not so important feature

Either we must predict price or price per sqft. Because both are interdependent

* Also drop availability and area_type

```python
type_average = get_mean_price(df,'type')
type_average
```

```python
housing_type_average = get_mean_price(df,'housing_type')
housing_type_average
```

```python
df['Location'].unique()
```

```python
location_average = get_mean_price(df,'Location')
location_average
```

* We need to assign weights to locations that would facilitate price prediction

```python
min_price = df['price'].min()-1 
max_price = df['price'].max()+1
```

```python
for loc, prices in location_average.items():
    prices = (prices-min_price)/(max_price-min_price)
    location_average[loc] = prices*100
```

```python
location_average
```

```python
df_BI = df.drop('price_sqft',axis=1)
df_BI.to_csv('Cleaned_data_analysis.csv')
```

```python
df['location_weight'] = df.apply(lambda x: location_average[x['Location']],axis=1)
```

```python
df.head()
```

```python
# Dropping less important features
df = df.drop(columns=['Location','price_sqft','area_type','availability'],axis=1)
df.head()
```

```python
df.to_csv('CleanedMumbaiHousing.csv')
```

```python
fig,ax = plt.subplots(nrows = 2,ncols=2,figsize=(20,10))
ax[0,0].bar(area_type_average.keys(),area_type_average.values())
ax[0,1].bar(availability_average.keys(),availability_average.values())
ax[1,0].bar(type_average.keys(),type_average.values())
ax[1,1].bar(housing_type_average.keys(),housing_type_average.values());

ax[0,0].set(title="Average Price wrt Area Type",ylabel="Average Price")
ax[0,1].set(title="Average Price wrt Availability",ylabel="Average Price")
ax[1,0].set(title="Average Price wrt Structure",ylabel="Average Price")
ax[1,1].set(title="Average Price wrt Property Type",ylabel="Average Price");
```

### Encoding

```python
dummies_housing = pd.get_dummies(df[['housing_type']],drop_first=True).astype(int)
dummies_type = pd.get_dummies(df[['type']],drop_first=True).astype(int)
```

```python
dummies_housing
```

```python
dummies_type
```

```python
df1 = pd.concat([df,dummies_housing,dummies_type],axis=1)
df1
```

```python
df1.drop(['type','housing_type'],axis=1,inplace=True)
```

```python
df1.head()
```

```python
X = df1.drop('price',axis=1)
y = df1.price
```

# Training

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=0,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf_reg = RandomForestRegressor()
rf_reg.fit(X_train,y_train)
```

```python
rf_reg.score(X_test,y_test)
```

Base Model Accuracy : 91.11 %

### Selecting the best Model

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import GradientBoostingRegressor
```

```python
models = {
    'RandomForestRegression':RandomForestRegressor(),
    'LinearRegression' : LinearRegression(),
    'GBS':GradientBoostingRegressor(),
    'KNN': KNeighborsRegressor()
}
```

```python
def get_best_model(models,X_train,X_test,y_train,y_test):
    model_score={}
    for model_name,model in models.items():
        clf = model.fit(X_train,y_train)
        score = clf.score(X_test,y_test)
        model_score[model_name] = score
    return model_score
```

```python
score_dict = get_best_model(models,X_train,X_test,y_train,y_test)
score_dict
```

```python
fig,ax = plt.subplots(figsize=(10,3))

ax.bar(score_dict.keys(),score_dict.values())
```

## Tuning the hyperparameters

```python
from sklearn.model_selection import GridSearchCV
```

```python
random_forest_params = {
    'n_estimators' : range(50,150,10),
    'min_samples_split': [2,4,6],
    'min_samples_leaf':[1,3]
}
```

```python
gs_rf_reg = GridSearchCV(RandomForestRegressor(),
                          param_grid=random_forest_params,
                          cv=5,
                          verbose=2)
gs_rf_reg.fit(X_train,y_train)
```

```python
gs_rf_reg.score(X_test,y_test)
```

```python
gs_rf_reg.best_params_
```
